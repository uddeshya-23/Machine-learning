```python
full_name={'N': 'Nabisco','Q': 'Quaker Oats','K': 'Kelloggs','R': 'Raslston Purina','G': 'General Mills' ,'P' :'Post' ,'A':'American Home Foods Products'}
full_mfr=[]
```

Linear Regression , Assign Dependent Variable & Independent Variable

```python
X=df.copy()
```

```python
X1= df.iloc[:, 3:16]
X1.columns
```

```python
Y=X.pop('rating')
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y=train_test_split(X1,Y, test_size=0.25, random_state=1001)
```

Random state for generating same record everytime

```python
train_x
```

```python
train_y, test_y
```

```python
#initialize the model
from sklearn.linear_model import LinearRegression
model_lr=LinearRegression()
```

```python
#fit data 
model_lr=model_lr.fit(train_x,train_y)
```

```python
train_x.columns
```

```python
model_lr.score(train_x, train_y)
```

```python
model_lr.score(test_x,test_y)
```

```python
predicted_ratings = model_lr.predict(test_x)
```

```python
#RMSE
import numpy as np
np.sqrt(sum((test_y-predicted_ratings)**2/len(test_y)))
```

```python
from sklearn.linear_model import Ridge
ridge_reg_1=Ridge(alpha=1,normalize=True)
ridge_reg_1.fit(train_x,train_y)
t=ridge_reg_1.score(test_x,test_y)   #alpha =1
ridge_reg_05=Ridge(alpha=0.5,normalize=True)
ridge_reg_05.fit(train_x,train_y)
t1=ridge_reg_05.score(test_x,test_y)   #alpha =0.5
ridge_reg_2=Ridge(alpha=2,normalize=True)
ridge_reg_2.fit(train_x,train_y)
t2=ridge_reg_2.score(test_x,test_y)    #alpha =2
```
